```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# 서로 다른 data frame을 concatenate하는 방법
python_df = pd.concat([country_respondents,country_uses_python], axis = 'columns', sort = False)
python_df
```

```python
python_df.rename(columns = {'Country':'NumRespondenst', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace = True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondenst']) * 100
```
